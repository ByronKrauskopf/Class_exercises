```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os
```

```python
# Read CSV
file_to_load = os.path.join("Resources", "got.csv")
got_df = pd.read_csv(file_to_load)
got_df
```

```python
# Get attacker and defender king data
attacker = got_df["attacker_king"].value_counts()
print(attacker)

defender = got_df["defender_king"].value_counts()
print(defender)
```

```python
# Get total battle data Series
total_battle = attacker.add(defender, fill_value=0)
total_battle
```

```python
# Configure plot and ticks
total_battle.plot(kind="bar", color="red")

# Set textual properties
plt.ylabel("Number of Battles")
plt.xlabel("Kings")
plt.title("Battle Kings")
plt.xticks(rotation=45)

# Show plot
plt.show()

# Resize plot to display labels
plt.tight_layout()
```

```python
#Create Dataframe
battle_df = pd.DataFrame({"attacker": attacker,
                         "defender": defender}).fillna(0)
battle_df
```

```python
#Pot stacked bar chart
battle_df.plot(kind="bar", stacked=True)
plt.ylabel("Number of Battles")
plt.xlabel("Kings")
plt.title("Battle Kings")
plt.xticks(rotation=45)
plt.tight_layout()
```
